```python
import re
import requests
import pandas as pd

from bs4 import BeautifulSoup
```

```python
# url='https://pl.youtubers.me/india/all/top-1000-youtube-channels-in-india/en'

url="https://pl.youtubers.me/india/all/top-1000-youtube-channels-in-india/en"
```

```python
url
```

```python
request_header= {'Content-Type':'text/css; charset=UTF-8',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36',
     'Accept-Encoding': 'gzip, deflate, br'
}
```

```python
response=requests.get(url,headers=request_header)
```

```python
print(response.status_code)
```

```python
response.headers
```

```python
html_code=response.text
```

```python
html_code
```

```python
soup=BeautifulSoup(html_code)
```

```python
soup
```

## Fretch Entire Data

```python
main_content=soup.find('div',attrs={'class':"main col col-sm-12"})
```

```python
main_content
```

## Fretch Sub data


## Fretch the Title

```python
sub_content=main_content.find('table',attrs={'class':'top-charts'})
```

```python
sub_content.text
```

```python
title = sub_content.find_all("a")
```

```python
data=sub_content.find_all('td')
```

```python
for i in data:
    print(i.text)
```

# Channel Names

```python
channel_names=[]
channels=[]

for i in title:
    channels.append(i.text)
    
a=channels[::2]
for i in a:
    channel_names.append(i)
```

```python
channel_names
```

```python
channels
```

#  Category

```python
category=channels[1::2]
```

```python
category
```

# Subscribers

```python
subscribers=[]
for i in data[2::7]:
    subscribers.append(i.text)
    
```

```python
subscribers
```

# video views

```python
views_counts=[]

for i in data[3::7]:
    views_counts.append(i.text)
```

```python
views_counts
```

# video count

```python
video_count=[]

for i in data[4::7]:
    video_count.append(i.text)
```

```python
video_count
```

```python
year=[]
for i in data[6::7]:
    year.append(i.text)
```

```python
Rank=[]
for i in data[::7]:
    Rank.append(i.text)
    
```

```python
print('channel_name',len(channel_names))
print('subscribers',len(subscribers))
print('category',len(category))
print('views_counts',len(views_counts))
print('video_count',len(video_count))
print('started_year',len(year))
print('rank',len(Rank))
```

```python
df=pd.DataFrame({
    'channel_name':channel_names,
    'subscribers':subscribers,
    'category':category,
    'views_counts':views_counts,
    'video_count':video_count,
    'started_year':year,
    'rank':Rank
})
```

```python
df.to_csv("webscrapping_data.csv",index=False)
```
